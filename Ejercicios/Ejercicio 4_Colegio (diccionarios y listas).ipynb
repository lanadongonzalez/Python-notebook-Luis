## Ejercicio: Colecciones colegio (diccionarios y listas)

En este ejercicio consideraremos un colegio con 3 aulas e implementaremos la siguiente funcionalidas:

1. Insertar alumno --> En qué aula quieres meter al alumno (1,2,3) --> DNI, Nombre, Apellidos
2. Mostrar alumnos colegio --> Todos los alumnos del colegio por aula.
3. Mostrar alumnos aula --> De qué aula quieres mostrar los alumnos (1,2,3)
4. Buscar alumno --> Preguntar DNI --> En que aula se encuentra el alumno ese DNI
5. Borrar alumno --> Preguntar DNI --> Borrar el alumno con ese DNI
6. Salir


```python
opcion = 0
colegio = [{'123456789Y':['Luis', 'Anadón', 'González'],'1234':['Eva', 'Martín', 'García']}
           ,{'4567':['luci','san', 'gon']},
           {'45678':['roci','berlanga', 'camus']}] #El colegio es una lista con tres diccionarios

while opcion != '6':
    print("¿Qué opción desea ejecutar? Por favor, escriba el número correspondiente\n 1. Insertar alumno\n 2. Mostrar alumnos colegio\n 3. Mostrar alumnos aula\n 4. Buscar alumno\n 5. Borrar alumno\n 6.Salir")
    opcion = input()
    
    if opcion == '1':
        print("Has elegido inserta un alumno. Rellene los siguientes datos")
        aula = input("Elija el aula del alumno (1, 2 o 3)")  #El usuario elige un aula
        
        if ((aula == '1') or (aula == '2') or (aula == '3')):  #Nos aseguramos que el aula es correcta
            dni = input("Escriba el DNI del alumno")    #El usuario elige los datos correspondientes
            nombre = input("Escriba el nombre del alumno")
            apellido = input("Escriba el primer apellido del alumno")
            apellido2 = input("Escriba el segundo apellido del alumno")
            colegio[int(aula)-1][dni] = [nombre, apellido, apellido2]    #Accedemos al aula deseada y metemos el usuario por la clave
            print("El alumno se ha introducido correctamente")
        else:
            print("Escoja un aula existente")
        
        
    elif opcion == '2':
        print("Estos son todos los alumnos del colegio:\n")
        for aula in colegio: #Recorro la lista
            for dni, alumno in aula.items(): #Recorro el diccionario
                if type(alumno)==list:     #Nos aseguramos que el alumno sea una lista
                    print(dni)
                    for i in range(len(alumno)):
                        print(alumno[i])
                else:
                    print(dni, alumno)
                print("----------")
            
    elif opcion == '3':
        aula = input("Elija el aula del alumno (1, 2 o 3)")

        if (aula == '1') or (aula == '2') or (aula == '3'):
            print("Estos son los alumnos del aula " + aula)
            for dni, alumno in colegio[int(aula)-1].items(): #Recorro el diccionario
                
                if type(alumno)==list:     #Nos aseguramos que el alumno sea una lista
                    print(dni)
                    for i in range(len(alumno)):
                        print(alumno[i])
                else:
                    print(dni, alumno)
                print("----------")
                
        else:
            print("Escoja un aula existente")
                   
                   
    elif opcion == '4':
        dni_alumno = input("Escriba el DNI del alumno")
        dni_no_encontrado = True #Variable para comprobar si hemos encontrado al alumno
        for aula in colegio:
            for dni, alumno in aula.items():
                if dni_alumno==dni:
                    dni_no_encontrado = False
                    if type(alumno)==list:     #Nos aseguramos que el alumno sea una lista
                        print(dni)
                        for i in range(len(alumno)):
                            print(alumno[i])
                    else:
                        print(dni, alumno)
        if dni_no_encontrado==True:
            print("El DNI introducido no está en nuestra base de datos")
        
    elif opcion == '5': 
        dni_alumno = input("Escriba el DNI del alumno que quiere eliminar")
        dni_no_encontrado = True #Variable para comprobar si hemos encontrado al alumno
        numero_aula_alumno = 0 #En este valor recogeremos el número de aula
        for numero_aula in range(len(colegio)):
            for dni in colegio[numero_aula].items():
                if dni_alumno==dni:
                    dni_no_encontrado = False
                    numero_aula_alumno = numero_aula
        if dni_no_encontrado==True:
            print("El DNI introducido no está en nuestra base de datos")
        else:
            del(aula[dni])
            print("El alumno ha sido borrado satisfactoriamente")
        
    elif opcion == '6': 
        print("Has seleccionado la opción 6. ¡Adios!")
    else:
        print("Elija una opción válida, por favor.")
```
