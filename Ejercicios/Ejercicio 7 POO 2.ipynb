```python
#Creamos la clase televisión
class Tv():
    def __init__(self, identificador, marca, pulgadas, precio):
        self.identificador = identificador
        self.marca = marca
        self.pulgadas = pulgadas
        self.precio = precio
```

```python
#Creamos la clase ordenador
class Ordenador():
    def __init__(self, identificador, marca, disco_duro, ram, precio):
        self.identificador = identificador
        self.marca = marca
        self.disco_duro = disco_duro
        self.ram = ram
        self.precio = precio
```

```python
#Creamos la clase monitor
class Monitor():
    def __init__(self, identificador, marca, pulgadas, precio):
        self.identificador = identificador
        self.marca = marca
        self.pulgadas = pulgadas
        self.precio = precio
```

```python
#Creamos la clase cliente
class Cliente():
    def __init__(self, identificador, nombre, vip = False, compra = []):
        self.identificador = identificador
        self.nombre = nombre
        self.vip = vip
        self.compra = compra
    
    def pagar(self, importe):
        print('El cliente {} está pagando un total de {}'.format(self.nombre, importe))
```

```python
#Creamos la clase cajero
class Cajero():
    def __init__(self, identificador, nombre):
        self.identificador = identificador
        self.nombre = nombre
    def cobrar(self, cliente):
        total = 0
        for producto in cliente.compra:
            total += producto.precio
        if cliente.vip:
            total = total*0.85 #Le hacemos un 15% de descuento si es VIP
 #       self.tienda.cliente.remove(cliente) #Una vez hemos cobrado, se elimina al cliente de la tienda
        return total
```
